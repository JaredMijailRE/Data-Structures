#### **Resumen de la Estructura de Datos de Árboles**  

Un **árbol** es una estructura de datos jerárquica que consta de **nodos** conectados por **aristas**. Se usa ampliamente en informática para representar relaciones jerárquicas como sistemas de archivos, estructuras organizacionales y árboles de decisión.  

#### **Elementos Claves de un Árbol**  
- **Raíz**: Es el nodo principal del árbol, sin padre.  
- **Nodo**: Elemento que puede contener un valor y enlaces a otros nodos.  
- **Hijo**: Nodo descendiente de otro nodo.  
- **Padre**: Nodo que tiene hijos.  
- **Hoja**: Nodo sin hijos.  
- **Altura**: Número de niveles desde la raíz hasta la hoja más profunda.  
- **Profundidad**: Distancia de un nodo hasta la raíz.  

#### **Tipos de Árboles**  
1. **Árbol Binario**: Cada nodo tiene como máximo dos hijos.  
   - **Árbol Binario de Búsqueda (BST)**: Organiza los nodos para facilitar la búsqueda, donde el hijo izquierdo es menor y el derecho mayor que el nodo padre.  
   - **Árbol Balanceado**: Mantiene una altura mínima para optimizar operaciones (ejemplo: AVL, Red-Black).  
2. **Árbol N-ario**: Cada nodo puede tener hasta *N* hijos.  
3. **Árbol de Expresión**: Representa expresiones matemáticas.  
4. **Árbol de Decisión**: Se usa en inteligencia artificial y aprendizaje automático.  
5. **Heap**: Árbol binario completo donde los nodos cumplen una propiedad de orden (ejemplo: Max-Heap, Min-Heap).  
6. **Trie**: Árbol utilizado en búsqueda eficiente de cadenas, útil en diccionarios y autocompletado.  

#### **Operaciones en Árboles**  
- **Recorridos**:  
  - *Preorden (Raíz - Izquierda - Derecha)*  
  - *Inorden (Izquierda - Raíz - Derecha)*  
  - *Postorden (Izquierda - Derecha - Raíz)*  
  - *Por niveles (BFS, usando una cola)*  
- **Búsqueda**: Encontrar un nodo específico.  
- **Inserción y Eliminación**: Modificar la estructura del árbol manteniendo sus propiedades.

#### BST

Esta estructura cuenta con height, insert, find, delete, traverse.


Primero vamos a crear la clase del bloque basico que vamos a usar para construir nuestros arboles.

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
```

Esta clase nodo avl va a tener una variable para el valor del nodo como tal, después vamos a tener dos variables que van a actuar como punteros a los nodos hijos, dado que este es un árbol binario, será el hijo izquierdo y derecho.

Ahora vamos a declarar el constructor de nuestra clase arbol BST.

```python
class BST:
    def __init__(self, value=None):
        """Inica el arbol, tiene un valor opcional."""
        self.root = None
        if value is not None:
            self.root = Node(value)
```

Al crear nuestro arbol vamos a poder darle un valor opcional a la hora de crear la raiz.

Siguiente realizaremos la funcion insert que nos permitira añadir nuevos nodos a nuestro arbol.

```python
def insert(self, value):
    """Insert a new node with the specified value."""
    if self.root is None:
        self.root = Node(value)
        return
    
    current = self.root
    while True:
        if value < current.value:
            if current.left is None:
                current.left = Node(value)
                return
            current = current.left
        elif value > current.value:
            if current.right is None:
                current.right = Node(value)
                return
            current = current.right
        else:
            return
```

En el algoritmo primero vemos si root es nulo, si lo es insertamos directamente el valor en la raiz, sino iniciamos un bucle while en el que buscaremos una hoja en la que podamos hacer la insercions, para ello hiremos de la raiz en adelante comparando si el vlaor que vamos a insertar es mayor o menor si este valor es menor entonces nos hiremos a la izquierda de lo contrario a la derecha.

El siguiente metodo es find, en el que dado un valor buscaremos un nodo con este valor.

```python

def find(self, value):
    """Search for a node with the given value.
    
    Returns:
        The node if found; otherwise, None.
    """
    current = self.root
    while current:
        if value < current.value:
            current = current.left
        elif value > current.value:
            current = current.right
        else:
            return current  
    return None
```

Pondremos en una variable nuestro root, y usaremos un algoritmo parecido al de insert en el cual dado con un while vamos moviendo nuestra variable entre nodos si es menor el valor que buscamos nos vamos a la izquierda si no lo es nos vamos a la derecha, si el valor no es menor ni mayor significa que es igual por lo que retornaremos el nodo en el que estamos, si no encontramos nada solo retornaremos nulo.

El siguiente metodo es el delete 

```python
def delete(self, value):
    """Delete the node with the given value from the tree."""
    def _delete(node, value):
        if node is None:
            return None
        if value < node.value:
            node.left = _delete(node.left, value)
        elif value > node.value:
            node.right = _delete(node.right, value)
        else:
            if node.left is None:
                return node.right
            if node.right is None:
                return node.left
            successor = node.right
            while successor.left:
                successor = successor.left
            node.value = successor.value
            node.right = _delete(node.right, successor.value)
        return node
    
    self.root = _delete(self.root, value)
```

Para eliminar nos encontramos con 3 casos posibles, primero debemos buscar el nodo que queremos eliminar de forma que tenemos un algoritmo parecido al de find, una ves tenemos el nodo puede que este no tenga hijos, que solo tenga un hijo o que tenga dos hijos.

De forma que dentro del metodo creamos una funcion recursiva, tenemos un primer if que verifica que la raiz del subarbol en el que estamos no sea nula, despues tenemos los 3 casos de buscada que sea menos nos vamos a la izquierda, si es mayor nos vamos a la derecha, y si no es ninguno entonces el nodo es el que buscamos eliminar.

ya en el nodo que buscamos, si no tiene hijo izquierdo, entonces devolvemos el derecho que remplazara el nodo que queremos eliminar. Si no tiene derecho, entonces realizaremos lo inverso. Ahora si tiene a los dos hijos, debemos buscar el sucesor in-order, osea el nodo mas paqueño en el subarbol derecho.

Para ello en un while vamos avanzando siempre a la izquierda, y una ves el nodo en el que estemos no tenga hijo izquierdo, ese es el nodo por el que remplazaremos el que queremos eliminar, y finalmente volvemos a llamar el delet para eliminar el nodo por el que lo estamos remplzando.

EL siguiente metodo sera height

```python
def height(self):
    """Return the height of the tree (number of levels)."""
    def _height(node):
        if node is None:
            return 0
        return 1 + max(_height(node.left), _height(node.right))
    return _height(self.root)
```

Esta tambien es una funcion recusiva, que basicamente toma todos los caminos posibles, y al volver a los llamados originales compara cual es el conteo que se lleva a la izquierda y a la derecha, quedandose solo con el mayor, osea el que tiene mayor profundidad.

finalmente tenemos la funcion traverse

```python
def traverse(self):
    """Return an in-order traversal (sorted order) of the tree's values."""
    result = []
    def _inorder(node):
        if node:
            _inorder(node.left)
            result.append(node.value)
            _inorder(node.right)
    _inorder(self.root)
    return result
```

Este metodo es uno de los 3 tipos de recorridos que se pueden realizar, en este caso es un in order, basicamente nos vamos al sub arbol mas a la izquierda cuando encontramos la hoja, añadirmos ese valor, despues añadimos el valor del nodo al que pertenece esa hoja, y despues el valor de la otra hoja, la derecha, y hacemos esto sucesivamente subiendo en el arbol, regresando de la funcion recursiva.

```python
class NodoAVL:
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None
        self.altura = 1
```

Iniciamos con nuestra clase nodo con la cual vamos a construir nuestro arbol, la definimos de forma similar a como definimos la anterior, sin embargo ahora necesitaremos el atributo altura que nos va a permitir mantener el balance de nuestro arbol.

```python
def _nodeHeight(self, nodo):
    if not nodo:
        return 0
    return nodo.altura

def height(self):
    return _nodeHeight(self.root)

def getBalance(self, nodo):
    if not nodo:
        return 0
    return self._nodeHeight(nodo.izquierdo) - self._nodeHeight(nodo.derecho)
```


En relacion con nuestro nuevo atributo tenemos estos dos nuevos metodos, que son height y getBalance, el primero simplemente dado un nodo accede a su atributo altura, osea es un getter del mismo, y tenemos get balance, que nos da el balance de un nodo, returna la altura de su nodo izquierdo - el derecho lo que sera conveniente.

```python
def rotacionDerecha(self, z):
    y = z.izquierdo
    T3 = y.derecho
    
    y.derecho = z
    z.izquierdo = T3
    
    z.altura = 1 + max(self.height(z.izquierdo), self.height(z.derecho))
    y.altura = 1 + max(self.height(y.izquierdo), self.height(y.derecho))
    
    return y

def rotacionIzquierda(self, z):
    y = z.derecho
    T2 = y.izquierdo
    
    y.izquierdo = z
    z.derecho = T2
    
    z.altura = 1 + max(self.height(z.izquierdo), self.height(z.derecho))
    y.altura = 1 + max(self.height(y.izquierdo), self.height(y.derecho))
    
    return y
```

Ahora tenemos las rotaciones, la izquierda y la derecha, que ralizan la rotacion mejor explicada atraves de una imagen

Ahora si tenemos el insert que primero se encarga de insertar el nodo a la izquierda o derecha, y despues de hacer el balanceo una ves insertado teniendo 4 casos posibles en los que va a ser necesario rezalizar una rotacion distinta.

#### Heap

Para el heap tenemos implementaciones mas simples

```python
def _sift_up(self, index):
    """
    Mueve el elemento en el índice hacia arriba hasta restaurar la propiedad del heap.
    """
    parent = (index - 1) // 2
    while index > 0 and self.heap[index] < self.heap[parent]:
        self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
        index = parent
        parent = (index - 1) // 2

def _sift_down(self, index):
    """
    Mueve el elemento en el índice hacia abajo hasta restaurar la propiedad del heap.
    """
    n = len(self.heap)
    while 2 * index + 1 < n:
        smallest = index
        left = 2 * index + 1
        right = 2 * index + 2

        if left < n and self.heap[left] < self.heap[smallest]:
            smallest = left
        if right < n and self.heap[right] < self.heap[smallest]:
            smallest = right

        if smallest == index:
            break

        self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
        index = smallest
```

Aca tenemos la parte mas compleja del heap que es el sift up y el sift down, que ya se explico en clase su funcionamienot y es mas facil de entender visualmente.

```python
def insert(self, value):
    """Inserta un elemento en el heap."""
    self.heap.append(value)
    self._sift_up(len(self.heap) - 1)
```

La funcion insert simplemente añade el valor a la lista y despues realiza un sift up para mantener las propiedades del heap

```python
def find(self, value):
    """
    Busca un elemento en el heap.
    Devuelve el índice del elemento si se encuentra, o -1 en caso contrario.
    """
    for i in range(len(self.heap)):
        if self.heap[i] == value:
            return i
    return -1
```

Para el find recorremos toda la lista comparando los valores del heap con el valor que buscamos.

```python
def delete(self, value):
    """
    Elimina un elemento del heap.
    Devuelve True si se eliminó el elemento, o False si no se encontró.
    """
    index = self.find(value)
    if index == -1:
        return False  # Elemento no encontrado

    last_index = len(self.heap) - 1
    # Reemplaza el elemento a eliminar con el último elemento
    self.heap[index] = self.heap[last_index]
    self.heap.pop()

    # Si no era el último elemento, reajustar el heap
    if index < len(self.heap):
        self._sift_down(index)
        self._sift_up(index)
    return True
```

Finalmente tenemos el delete con el que buscamos primero el valor que queremos eliminar y despues remplazamos el ultimo elemento por el elemento que queremos eliminar, despues re ajustamos la lista para que mantenga su propiedad de heap

```python
def medir_para_estructura(estructura_tipo, test_samples, num_elementos):
    """Mide y acumula los tiempos de inserción para la estructura indicada en todos los tests."""
    tiempos_total = [0] * num_elementos
    for valores in test_samples:
        if estructura_tipo == 'avl':
            estructura = avl.ArbolAVL()
        elif estructura_tipo == 'bst':
            estructura = bst.BST()
        elif estructura_tipo == 'heap':
            estructura = heap.Heap()
        else:
            raise ValueError("Tipo de estructura desconocido")
        
        tiempos = medir_tiempos_insercion(estructura, valores)
        # Acumula los tiempos de cada inserción
        for i in range(num_elementos):
            tiempos_total[i] += tiempos[i]
    return tiempos_total

# Parámetros de la prueba
num_elementos = 2000
num_tests = 2000

test_samples = [random.sample(range(num_elementos * 10), num_elementos) for _ in range(num_tests)]

# Creamos un pool con 3 procesos, uno para cada estructura.
pool = multiprocessing.Pool(processes=3)

result_avl = pool.apply_async(medir_para_estructura, args=('avl', test_samples, num_elementos))
result_bst = pool.apply_async(medir_para_estructura, args=('bst', test_samples, num_elementos))
result_heap = pool.apply_async(medir_para_estructura, args=('heap', test_samples, num_elementos))

# Recogemos los resultados de cada proceso.
tiempos_avl_total = result_avl.get()
tiempos_bst_total = result_bst.get()
tiempos_heap_total = result_heap.get()

pool.close()
pool.join()

# Calculamos los tiempos promedio por operación de inserción.
tiempos_avl_promedio = [t / num_tests for t in tiempos_avl_total]
tiempos_bst_promedio = [t / num_tests for t in tiempos_bst_total]
tiempos_heap_promedio = [t / num_tests for t in tiempos_heap_total]

# Graficar los resultados
plt.figure(figsize=(10, 6))
# Se grafican solo los primeros num_elementos-100 puntos para facilitar la visualización.
plt.plot(range(num_elementos-100), tiempos_avl_promedio[:num_elementos-100], label="AVL")
plt.plot(range(num_elementos-100), tiempos_bst_promedio[:num_elementos-100], label="BST")
plt.plot(range(num_elementos-100), tiempos_heap_promedio[:num_elementos-100], label="Heap")
plt.xlabel("Número de operación de inserción")
plt.ylabel("Tiempo de inserción promedio (segundos)")
plt.title("Tiempo de inserción promedio para AVL, BST y Heap")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def fillData(estructura, valores):
    for val in valores:
        estructura.insert(val)

def medir_para_estructura(estructura_tipo, test_samples, num_elementos):
    """Mide y acumula los tiempos de inserción para la estructura indicada en todos los tests."""
    tiempos_total = [0] * num_elementos
    for valores in test_samples:
        if estructura_tipo == 'avl':
            estructura = avl.ArbolAVL()
        elif estructura_tipo == 'bst':
            estructura = bst.BST()
        elif estructura_tipo == 'heap':
            estructura = heap.Heap()
        tiempos = medir_tiempos_delete(estructura, valores)
        # Acumula los tiempos de cada inserción
        for i in range(num_elementos):
            tiempos_total[i] += tiempos[i]
    return tiempos_total

def medir_tiempos_delete(estructura, valores):
    fillData(estructura, valores)
    tiempos = []
    for valor in valores:
        inicio = time.perf_counter()
        estructura.delete(valor)
        fin = time.perf_counter()
        tiempos.append(fin - inicio)
    return tiempos
```

### Heap con N hijos

Solo nos faltaba el heap con N hijos para los cuales generaremos de igual forma 3 graficas correspondiente a cada metodo, y analizaremos hasta 12 hijos, haremos el analizis de 2 en dos para poder ver mejor la diferencia entre los mismos.

```python
from Trees import heapN
```

```python
def medir_para_heap_d(d, test_samples, num_elementos, medir_tiempos):
    """
    Para un heap d-ario (con aridad d), ejecuta num_tests tests (cada uno con num_elementos)
    y acumula los tiempos de inserción para cada operación.
    """
    tiempos_total = [0] * num_elementos
    for valores in test_samples:
        # Se crea una nueva instancia del heap para cada test
        estructura = heapN.DaryHeap(d)
        tiempos = medir_tiempos(estructura, valores)
        for i in range(num_elementos):
            tiempos_total[i] += tiempos[i]
    return tiempos_total

def getResult(medir_tiempos, num_elementos, num_tests, d_values):
    # Parámetros de la prueba
    num_elementos = num_elementos  # Número de inserciones por test
    num_tests = num_tests       # Cantidad de tests a ejecutar

    # Generamos una lista de muestras de prueba: cada test es una lista con num_elementos valores aleatorios
    test_samples = [random.sample(range(num_elementos * 10), num_elementos) for _ in range(num_tests)]
    
    # Usamos un pool de procesos: se crea un proceso para cada valor de d
    pool = multiprocessing.Pool(processes=len(d_values))
    async_results = {}
    
    for d in d_values:
        async_results[d] = pool.apply_async(medir_para_heap_d, args=(d, test_samples, num_elementos, medir_tiempos))
    
    pool.close()
    pool.join()
    
    return async_results

```

Aca realizamos una optimizacion respecto al codigo pues ahora definimos una sola funcion que realiza los experimentos solo teniendo que cambiar la funcion del experimento como tal

#### Python multiprocessing

Un problema con Python es el GIL, que es el Global Interpreter Lock, que hace difícil tener paralelismo en el mismo, aunque quitarlo es algo que se está probando en la versión más reciente de Python. Por otro lado, actualmente existen maneras de tener varios procesos en Python; una de estas es multiprocessing, básicamente primero uno define un pool con una cantidad de procesos, estas son instancias de Python que pueden ejecutarse por separado, como correr varios programas de Python al mismo tiempo, con un intérprete por separado, como procesos separados. Una vez creado el pool se pueden asignar múltiples tareas que cada worker de este debe completar, cada worker correspondería a cada instancia de Python por separado que creamos, en este caso 3. Después, con get, obtenemos los resultados de cada proceso de Python, y después tenemos que cerrar los procesos y hacerles join, para acabar con los procesos que teníamos, liberar la memoria que estos estaban usando y volver al programa de Python principal.

Hay otras librerías que nos permiten hacer cosas similares, como threading que utiliza hilos dentro del mismo proceso, por lo que se comparte el mismo espacio de memoria; por otro lado, esto no es paralelismo real, pues siguen siendo tareas dentro del mismo proceso, finalmente está async que utiliza el mismo hilo pero evita el bloqueo por tareas, el término clave en estos dos casos es concurrencia, solo que en un caso es con varios hilos y en este último en un solo hilo.

Cabe decir que un aspecto negativo de multiprocessing es que el crear estos nuevos procesos de Python también conlleva su complejidad, por esto no es que veamos que los tiempos sean 3 veces más rápidos, sin embargo podríamos notar más los beneficios mientras más largas sean las tareas que se tengan que ejecutar.
